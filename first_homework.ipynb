# Уменьшение количества признаков

Как видно из таблицы, некоторые значения в различных пикселях коррелируют друг с другом. Для того, чтобы объединить наиболее зависимые признаки, можно использовать методы факторного анализа. Также благодаря этому можно уменьшить размерность данных. Это позволит проводить более наглядный анализ.

Одним из методов многомерного факторного анализа является `метод главных компонент`. Одной из областей, в которой данный метод часто используется, является обработка изображений.

Одна из постановок задачи для данного метода заключается в нахождении таких линейных многообразий (главных компонент), для которых разброс проекции исходнных данных будет максимален. 

Метод применяется только для количественных признаков, поэтому сначала необходимо отделить метки классов от значений пикселей.

```python
X = data.drop(('label'), axis = 1)
y = data['label']
print(X.shape)
print(y.shape)
```

Теперь можно производить редукцию. Необходимо выбрать количесвто главных компонент (новое количество признаков). Для начала возьмем произвольное число, пусть оно будет равно шести.

```python
from sklearn.decomposition import PCA
```

```python
n_components = 6
pca = PCA(n_components=n_components)
pca.fit(X)
reduced_X = pca.transform(X)
```

```python
reduced_X.shape
```

```python
print(type(X))
print(type(reduced_X))
```

После редукции у данных поменялся тип. Поэтому для удобства дальнейшей обработки лучше конвертировать их обратно в `Pandas.DataFrame`.

```python
# Convert to pandas
feature_dict = {}
for i in range(n_components):
    feature_name = 'principal_component_{}'.format(i + 1)
    feature_dict[feature_name] = reduced_X[:, i]

reduced_X = pd.DataFrame(feature_dict)
```

```python
reduced_X.describe()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(reduced_X, alpha = .01, figsize = (15, 15))
pass
```

```python
reduced_X.corr()
```

Как видно из таблицы, корреляция между различными признаками практически равна нулю. Этого мы и добивались. Теперь можно задаться следующим вопросом: 

А сколько информации было потеряно при таком сжатии? И какое количество главных компонент нужно выбрать, чтобы сохранить требуемое количество информации.

```python
percents = np.arange(0.5, 1.0, 0.05)
components = []
for data_safety in percents:
    safety_pca = PCA(data_safety)
    new_data = safety_pca.fit_transform(X)
    components.append(safety_pca.components_.shape[0])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (10, 8))
ax.plot(components, percents * 100)
ax.set_title('Метод главных компонент')
ax.set_xlabel('Новое количество признаков')
ax.set_ylabel('Количесвто сохраненной информации (%)')
```

Из графика видно, что после 80% сохраненной информации на каждый новый процент требутеся добавлять много дополнительных главных компонент, и получается, что в методе практически нет смысла.

Поэтому возьмем в качестве нового количества признаков 25 главных компонент, что позволит сжать исходное количество признаков примерно в 30 раз, потеряв при этом всего 20% исходной информации.

```python
from sklearn.decomposition import PCA

n_components = 25
pca = PCA(n_components=n_components)
pca.fit(X)
X = pca.transform(X)

# Convert to pandas
feature_names = []
feature_dict = {}
for i in range(n_components):
    feature_name = 'principal_component_{}'.format(i + 1)
    feature_names.append(feature_name)
    feature_dict[feature_name] = X[:, i]

X = pd.DataFrame(feature_dict)
```

# Классификация

## Разделение данных на обучающую и тестовую выборки

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

```python
knn
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

```python
# Подбор параметров
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```
